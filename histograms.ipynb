**HISTOGRAM FOR 2D NUMPY ARRAY**

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(20, 40, size=(10, 2))
plt.hist(data, bins=5)
plt.xlabel("Array values")
plt.ylabel("Frequency")

x_ticks = np.arange(20, 41, 2)
plt.xticks(x_ticks)

y_ticks = np.arange(0, 6, 1)
plt.yticks(y_ticks)

print(data)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000, 2)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].hist(data[:, 0], bins=20, alpha=0.5, color='blue', label='Column 1')
axes[1].hist(data[:, 1], bins=20, alpha=0.5, color='green', label='Column 2')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Histogram of Column 1')
axes[0].legend()
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of Column 2')
axes[1].legend()
plt.show()
```
